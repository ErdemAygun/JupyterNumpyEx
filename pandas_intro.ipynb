```python
data = pd.read_csv("titanic.csv", dtype={"Pclass": np.int8})  # using dtype parameter we can specify data type for any column
data
```

```python
data.info()
```

```python
data["Survived"] = data["Survived"].astype(bool)
```

```python
data
```

```python
data["Sex"] = data["Sex"].astype("category")
data.info()
```

```python
data
```

```python
x = data.head(10)  # select first 10 rows
x.to_excel("first_rows.xlsx")
```

```python
# To be able to select rows we need to use `loc` or `iloc` attribute
data.loc[0] # loc gives us access to rows by their index
```

```python
last = data.tail(10) # select last 10 rows
last.loc[881]  # OK
last.iloc[0]  # .iloc gives access to rows by their position (iloc[0] will always return the first row)
last.iloc[1:5]
```
